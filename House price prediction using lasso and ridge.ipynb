# Correlation

```python
numeric = data.select_dtypes(include=[np.number])
```

```python
categorical= data.select_dtypes(include=[np.object])
```

```python
from sklearn.preprocessing import OrdinalEncoder
OE= OrdinalEncoder()
for i in categorical.columns:
    data[[i]]=OE.fit_transform(data[[i]])
```

```python
correlation=numeric.corr()
print(correlation['SalePrice'].sort_values(ascending=False),'\n')
```

```python
f,ax=plt.subplots(figsize=(10,8))
sns.heatmap(correlation,square=True, vmax=0.8)
```

```python
k = 11
cols = correlation.nlargest(k,'SalePrice')['SalePrice'].index
print(cols)
cm = np.corrcoef(data[cols].values.T)
f, ax = plt.subplots(figsize=(14,12))
sns.heatmap(cm, vmax=0.8, linewidths=0.01,square=True,annot=True,cmap='viridis',
            linecolor="white",xticklabels= cols.values, annot_kws={'size':12},yticklabels = cols.values)
```

# Removal of outliers

```python
data["SalePrice"].describe()
```

```python
sns.boxplot(x=data["SalePrice"])
```

```python
data.shape
```

```python
first_quartile=data["SalePrice"].quantile(.25)
third_quartile=data["SalePrice"].quantile(.75)
IQR=third_quartile-first_quartile
```

```python
new_boundary=third_quartile+3*IQR
```

```python
data.drop(data[data['SalePrice']>new_boundary].index,axis=0,inplace=True)
```

```python
data.shape
```

```python
data.fillna(method='bfill',inplace=True)
```
